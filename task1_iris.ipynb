```python
# Task 1 - Iris Dataset
# Step 1: Libraries Import

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Sab libraries ready hain! ✅")
```

```python
# Step 2: Dataset Load karo
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

print("Dataset shape:", df.shape)
print("\nPehli 5 rows:")
df.head()
```

```python
# Step 3: Dataset ki Info aur Statistics
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nDataset Info:")
print(df.info())
print("\nStatistics:")
df.describe()
```

```python
# Step 4: Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', 
                hue='species', palette='Set1')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
# Step 5: Histogram
df.drop('species', axis=1).hist(figsize=(10,6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Box Plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df, palette='Set2')
plt.title('Box Plot - Outliers Dekhne ke Liye')
plt.xticks(rotation=15)
plt.show()
```

```python
# Step 7: Final Insights
print("=== TASK 1 - FINAL INSIGHTS ===")
print(f"\nTotal Flowers: {len(df)}")
print(f"Total Features: {len(df.columns)-1}")
print(f"\nSpecies Count:")
print(df['species'].value_counts())
print("\nKoi missing values hain? :", df.isnull().sum().sum())
print("\nConclusion: Iris dataset mein 3 species hain.")
print("Setosa baaki donon se clearly alag hai!")
```
